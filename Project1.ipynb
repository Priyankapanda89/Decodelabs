Import Pandas


```python
import pandas as pd
```

Load Dataset

```python
df = pd.read_excel("/content/Dataset for Data Analytics (1).xlsx")
```

```python
df.head()
```

```python
df.info()
```

check data set shape

```python
print(df.shape)
```

Check missing values

```python
print(df.isnull().sum())
```

Handle missing values

```python
df["CouponCode"] = df["CouponCode"].fillna("NO COUPON")
```

```python
print(df["CouponCode"].isnull().sum())
```

Check Duplicates Rows

```python
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
print(df.duplicated().sum())
```

Check duplicate OrderIDs

```python
print(df["OrderID"].duplicated().sum())
```

Clean Text Fields

```python
for col in df.select_dtypes(include = "object"):
  df[col] = df[col].str.strip()
```

Verify Date format

```python
df["Date"] = pd.to_datetime(df["Date"],errors ="coerce")
```

```python
print(df["Date"].isnull().sum())
```

check final Data quality

```python
print("Missing Values")
print(df.isnull().sum())
print("\nDuplicate Rows")
print(df.duplicated().sum())
print("\nDuplicated OrderIDs")
print(df["OrderID"].duplicated().sum())
print("\nInvalid Dates")
print(df["Date"].isnull().sum())
```

```python
df.to_excel("/content/Dataset for Data Analytics (1).xlsx", index = False)
```
